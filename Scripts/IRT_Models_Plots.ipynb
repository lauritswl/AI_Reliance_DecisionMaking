```python
import numpy as np
import matplotlib.pyplot as plt

# Define the logit function
def logit(a):
    return 1 / (1 + np.exp(-a))

# Person ability (log odds) from -6 to 6
a = np.linspace(-6, 6, 1000)

# Apply logit function to person ability
probability = logit(a)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(a, probability, label="Logistic Function")
plt.axhline(0.7310586, color='r', linestyle='--', label="Probability = .73")
plt.axvline(1, color='g', linestyle='--', label="Person Ability (a[i] = 1)")
plt.title("Item Response Function: P(X[i] = 1 | a[i])")
plt.xlabel("Person Ability (a[i])")
plt.ylabel("P(X[i] = 1)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the logit function with d_j as a parameter
def logit(a, d):
    return 1 / (1 + np.exp(-(a - d)))

# Person ability (log odds) from -6 to 6
a = np.linspace(-10, 10, 1000)

# Define different values for d_j
d_values = [-2.5, 0, 5]
colors = ['green', 'blue', 'red']  # Specify the colors for each d_j

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the models for different d_j values
for i, d in enumerate(d_values):
    probability = logit(a, d)
    plt.plot(a, probability, color=colors[i], label=f"d = {d}")
    plt.vlines(x=d, ymin=-0.03, ymax=0.5, linestyle='--', color=colors[i], linewidth=0.8)

plt.axhline(0.5, color='black', linestyle='-', label="Probability = .5")

plt.title("Item Response Function: P(X[i] = 1 | a[i], d[j])")
plt.xlabel("Person Ability (a[i])")
plt.ylabel("[ij] = P(X[i] = 1)")
plt.legend()
plt.grid(False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the logit function with d_j and s as parameters
def logit(a, d, s):
    return 1 / (1 + np.exp(-s * (a - d)))

# Person ability (log odds) from -10 to 10
a = np.linspace(-10, 10, 1000)

# Define different values for d_j and s
d_values = [-2.5, 0, 5]
s_values = [0., 1, 3]  # Sensitivity values
colors = ['green', 'blue', 'red']  # Specify the colors for each pair

# Create the plot
plt.figure(figsize=(10, 6))
# Plot the models for each pair of d_j and s
for d, s, color in zip(d_values, s_values, colors):
    probability = logit(a, d, s)
    plt.plot(a, probability, color=color, label=f"d={d}, s={s}")
    plt.vlines(x=d, ymin=-0.03, ymax=0.5, linestyle='--', color=color, linewidth=0.8)  # Vertical line stopping at y=0.5 slightly below 0,

plt.axhline(0.5, color='black', linestyle='-', label="Probability = 0.5")

plt.title("Item Response Function: P(X[i] = 1 | a[i], d[j], s[j])")
plt.xlabel("Person Ability (a[i])")
plt.ylabel("[ij] = P(X[i] = 1)")
plt.legend()
plt.grid(False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the logit function with d_j and s as parameters
def logit(a, d, s):
    return 1 / (1 + np.exp(-s * (a - d)))

# Person ability (log odds) from -6 to 6
a = np.linspace(-6, 6, 1000)

# Define different values for d_j and s
d_values = [0]  # Threshold at d = 1 for simplicity
s_values = [1]  # Sensitivity value
colors = ['blue']  # Color for the plot

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the models for each pair of d_j and s
for d, s, color in zip(d_values, s_values, colors):
    probability = logit(a, d, s)
    plt.plot(a, probability, color=color, label="Logistic Function")
    plt.vlines(x=d, ymin=0, ymax=0.5, linestyle='--', color=color, linewidth=0.8, label="Person Ability (a[i] = 0)")  # Vertical line stopping at y=0.5

# Add horizontal line for a specific probability
plt.axhline(0.5, color='black', linestyle='-', label="Probability = .5")

# Title and labels
plt.title("Item Response Function: P(X[i] = 1 | a[i])")
plt.xlabel("Person Ability (a[i])")
plt.ylabel("[ij] = P(X[i] = 1)")

# Display the legend
plt.legend()

# Enable grid
plt.grid(False)

# Show the plot
plt.show()
```
